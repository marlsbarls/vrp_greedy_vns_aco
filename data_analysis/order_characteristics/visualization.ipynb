```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
```

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.savefig(title)
    plt.show()
    
```

# Anzahl Orders pro Tag

```python
orders_per_day = pd.read_csv('ordersPerDay.csv')
mean = orders_per_day['count'].mean()
median = orders_per_day['count'].median()
print(f'Mittelwert: {mean}')
print(f'Median: {median}')
```

## Zeitreihe

```python
plot_df(orders_per_day, x=orders_per_day['date'], y=orders_per_day['count'], title='orders per day') 
```

## Boxplot

```python
bp = orders_per_day.boxplot(column=['count'])
# plt.savefig('bp_orders_per_day')
```

# Dynamische Orders pro Tag

```python
dynamic_per_day = pd.read_csv('dynamic_orders_per_day.csv')
mean = dynamic_per_day['count'].mean()
median = dynamic_per_day['count'].median()
print(f'Mittelwert: {mean}')
print(f'Median: {median}')
```

```python
dynamic_and_total_df = orders_per_day.copy()
dynamic_and_total_df = dynamic_and_total_df.rename(columns={'count': 'total_orders'})
dynamic_and_total_df['dynamic_abs'] = NaN

for index, row in dynamic_and_total_df.iterrows():
    if row.date in dynamic_per_day['date'].tolist():
        value = dynamic_per_day.loc[dynamic_per_day['date'] == row.date, 'count'].values[0]
        dynamic_and_total_df.loc[dynamic_and_total_df['date'] == row.date, ['dynamic_abs']] = value
    else:
        dynamic_and_total_df.loc[dynamic_and_total_df['date'] == row.date, ['dynamic_abs']] = 0

dynamic_and_total_df['dynamic_rel'] = dynamic_and_total_df['dynamic_abs']/dynamic_and_total_df['total_orders']
```

## Boxplot

```python
bp = dynamic_and_total_df.boxplot(column=['dynamic_rel'])
plt.savefig('dynamic_rel_per_day')

        
```

## Zeitreihe

```python
plot_df(dynamic_and_total_df, x=dynamic_and_total_df['date'], y=dynamic_and_total_df['dynamic_rel'], title='dynamic orders relative orders per day') 
```

# Cancellations

```python
cancellations_per_day = pd.read_csv('cancellations_per_day.csv')
mean = cancellations_per_day['count'].mean()
median = cancellations_per_day['count'].median()
print(f'Mittelwert: {mean}')
print(f'Median: {median}')
```

```python
cancellations_and_total_df = orders_per_day.copy()
cancellations_and_total_df = cancellations_and_total_df.rename(columns={'count': 'total_orders'})
cancellations_and_total_df['cancellations_abs'] = NaN

for index, row in cancellations_and_total_df.iterrows():
    if row.date in cancellations_per_day['date'].tolist():
        value = cancellations_per_day.loc[cancellations_per_day['date'] == row.date, 'count'].values[0]
        cancellations_and_total_df.loc[cancellations_and_total_df['date'] == row.date, ['cancellations_abs']] = value
    else:
        cancellations_and_total_df.loc[cancellations_and_total_df['date'] == row.date, ['cancellations_abs']] = 0

cancellations_and_total_df['cancellations_rel'] = cancellations_and_total_df['cancellations_abs']/cancellations_and_total_df['total_orders']
```

## Boxplot

```python
bp = cancellations_and_total_df.boxplot(column=['cancellations_rel'])
plt.savefig('bp_cancellations_rel_per_day')
```

## Zeitreihe

```python
plot_df(cancellations_and_total_df, x=cancellations_and_total_df['date'], y=cancellations_and_total_df['cancellations_rel'], title='cancellations relative orders per day') 

```
